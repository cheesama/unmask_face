```python
from PIL import Image
from torchvision import transforms
from torchvision.transforms.functional import to_pil_image

import requests
import torch
import torch.nn as nn
```

```python
img = Image.open(requests.get('https://image.similarpng.com/very-thumbnail/2020/09/A-man-wearing-medical-mask-on-transparent-background-PNG.png', stream=True).raw)
img
```

```python
transform = transforms.Compose([
                transforms.Resize((256,256)),
                transforms.ToTensor(),
                transforms.Normalize([0.5,0.5,0.5],[0.5,0.5,0.5]),
            ])

norm_img = transform(img)
norm_img
```

```python
tanh_img = nn.Tanh()(norm_img)
tanh_img
```

```python
torch.max(tanh_img)
```

```python
#denorm_img = to_pil_image(tanh_img * 0.5 + 0.5)
denorm_img = to_pil_image(norm_img * 0.5 + 0.5)
denorm_img
```
